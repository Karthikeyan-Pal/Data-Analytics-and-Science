```python
#Importing Pandas library
import pandas as pd
```

```python
#Importing category_sales.csv file & storing it in df1
df1=pd.read_csv(r'C:\Users\karth\Desktop\practice_problem_Fall_2019\category_sales.csv')
df1
```

```python
#Importing date_dim.csv file & storing it in df2
df2=pd.read_csv(r'C:\Users\karth\Desktop\practice_problem_Fall_2019\date_dim.csv')
df2
```

```python
#Importing store_lookup.csv file & storing it in df3
df3=pd.read_csv(r'C:\Users\karth\Desktop\practice_problem_Fall_2019\store_lookup.csv')
df3
```

```python
#merging date_dim.csv file dataframe & storing it in df1
df1=pd.merge(df1,df2,on='month_key',how='inner')
df1
```

```python
#merging store_lookup.csv file dataframe & storing it in df1
df1=pd.merge(df1,df3,on='str_id',how='inner')
df1
```

```python
#Counting number of missing values in running_sales
df1.isnull().sum() 
```

```python
#Replacing Missing values in running_sales with average value
df1['running_sales']=df1['running_sales'].fillna(df1.groupby('str_id')['running_sales'].transform('mean'))
df1
```

```python
#Checking number of missing values in running_sales
df1.isnull().sum()
```

```python
#printing summary table
df4=df1.groupby(['channel_rollup','rtl_yr']).agg({'running_sales':['mean','sum']})
df4.columns=['avg_running_sales','tot_running_sales']
df4=df4.reset_index()
pd.options.display.float_format = '{:.2f}'.format
df4
```
