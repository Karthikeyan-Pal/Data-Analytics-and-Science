```python
#Importing Pandas & Numpy
import pandas as pd
import numpy as np
```

```python
#Importing Loan.csv file & storing it in df
df=pd.read_csv('C:/Users/karth/Desktop/Github/Loan Report-Python/Loan.csv')
df
```

```python
#Checking if LOAN_NUMBER is unique
boolean = df['LOAN_NUMBER'].duplicated().any()
boolean
```

```python
#Getting name of each column in the dataframe
list(df.columns.values)
```

```python
#Getting information about datatypes of each column
df.info()
```

```python
#Converting current balance into numeric data type
df[' CURRENT_BALANCE '] = df[' CURRENT_BALANCE '].str.replace('$', '')
df[' CURRENT_BALANCE '] = df[' CURRENT_BALANCE '].str.replace(',', '')
df[' CURRENT_BALANCE '] = pd.to_numeric(df[' CURRENT_BALANCE '])
```

```python
#Grouping data by Institution types & aggregating 'Loan Count','Average Current Balance','Max Current Balance','Min Current Balance' to create statistical report 1
df.rename(columns={'LENDER_INST_TYPE_DESCRIPTION':'Institution types'},inplace=True)
report1=df.groupby(['Institution types']).agg({'LOAN_NUMBER':'count',' CURRENT_BALANCE ':['mean','max','min']})
report1.columns=['Loan Count','Average Current Balance','Max Current Balance','Min Current Balance']
report1.at['Total','Loan Count']=report1['Loan Count'].sum()
report1.at['Total','Average Current Balance']=report1['Average Current Balance'].sum()
report1.at['Total','Max Current Balance']=report1['Max Current Balance'].sum()
report1.at['Total','Min Current Balance']=report1['Min Current Balance'].sum()
pd.options.display.float_format = '{:.2f}'.format
report1
```

```python
#Exporting report1 to a csv file
report1.to_csv('Report1.csv')
```

```python
#Creating statistical report 2
LTV = df[['LTV','LOAN_NUMBER',' CURRENT_BALANCE ']]
LTV.head()
```

```python
conditions = [(LTV['LTV'] <= 85.000000),(LTV['LTV'] > 85.000000) & (LTV['LTV'] <= 90.000000),(LTV['LTV'] > 90.000000) & (LTV['LTV'] <= 95.000000),(LTV['LTV'] > 95.000000)]

#Create a list of the values to assign for each condition
values = ['(<=85)', '(>85 & <=90)', '(>90 & <=95)', '(>95)']

#Create a new column and use np.select to assign values to it using our list
LTV['Loan_to_value_cohorts'] = np.select(conditions, values)

LTV.head()
```

```python
#Grouping data by Loan_to_value_cohorts & aggregating 'Loan Count','Average Current Balance','Max Current Balance','Min Current Balance' to create statistical report 2
report2=LTV.groupby(['Loan_to_value_cohorts']).agg({'LOAN_NUMBER':'count',' CURRENT_BALANCE ':['mean','max','min']})
report2.columns=['Loan Count','Average Current Balance','Max Current Balance','Min Current Balance']
report2.at['Total','Loan Count']=report2['Loan Count'].sum()
report2.at['Total','Average Current Balance']=report2['Average Current Balance'].sum()
report2.at['Total','Max Current Balance']=report2['Max Current Balance'].sum()
report2.at['Total','Min Current Balance']=report2['Min Current Balance'].sum()
pd.options.display.float_format = '{:.2f}'.format
report2
```

```python
#Exporting report2 to a csv file
report2.to_csv('Report2.csv')
```

```python
#Creating statistical report 3
LAC = df[['START_DATE','LOAN_ORIG_DATE','LOAN_NUMBER',' CURRENT_BALANCE ']]
LAC.head()
```

```python
#Calculating the difference between START_DATE and LOAN_ORIG_DATE 
#Storing the difference as a column in LAC dataframe
LAC['START_DATE'] = pd.to_datetime(df['START_DATE'])
LAC['LOAN_ORIG_DATE'] = pd.to_datetime(df['LOAN_ORIG_DATE'])                           
LAC['difference'] = LAC['START_DATE']-LAC['LOAN_ORIG_DATE']
LAC['difference'] = LAC['difference']/np.timedelta64(1,'M')
LAC['difference']=LAC['difference'].round()
LAC.head()
```

```python
conditions = [
              (LAC['difference'] !=0.0 ) & (LAC['difference'] <=0.0),
    
    (LAC['difference'] >= 0.0) & (LAC['difference'] <=9.0),
    (LAC['difference'] >= 10.0) & (LAC['difference'] <= 19.0),
    (LAC['difference'] >= 20.0) & (LAC['difference'] <= 29.0),
    (LAC['difference'] >= 30.0) & (LAC['difference'] <= 39.0),
    (LAC['difference'] >=40.0)
    ]

#Create a list of the values we want to assign for each condition
values = ['unknown','0-9 months', '10-19 months', '20-29 months','30-39 months','>=40 months']

#Create a new column and use np.select to assign values to it using our lists as arguments
LAC['loan_age_cohorts'] = np.select(conditions, values)

#Display updated DataFrame
LAC.head()
```

```python
#Grouping data by Loan_to_value_cohorts & aggregating 'Loan Count','Average Current Balance','Max Current Balance','Min Current Balance' to create statistical report 3
report3=LAC.groupby(['loan_age_cohorts']).agg({'LOAN_NUMBER':'count',' CURRENT_BALANCE ':['mean','max','min']})
report3.columns=['Loan Count','Average Current Balance','Max Current Balance','Min Current Balance']
report3.at['Total','Loan Count']=report3['Loan Count'].sum()
report3.at['Total','Average Current Balance']=report3['Average Current Balance'].sum()
report3.at['Total','Max Current Balance']=report3['Max Current Balance'].sum()
report3.at['Total','Min Current Balance']=report3['Min Current Balance'].sum()
pd.options.display.float_format = '{:.2f}'.format
report3
```

```python
#Exporting report3 to a csv file
report3.to_csv('Report3.csv')
```
